```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.neighbors import NearestNeighbors
import pickle
df = pd.read_csv("books.csv")
df.drop('Awards',axis=1,inplace=True)

allowed_features = ['Format', 'Theme (Color Style)', 'Genre','Language','Age Rating']
title_col = 'Title'
X_books = df[allowed_features]

categorical_cols = X_books.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X_books.select_dtypes(include=['int64', 'float64']).columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', StandardScaler(), numerical_cols)
    ]
)
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('knn', NearestNeighbors(n_neighbors=5, metric='cosine'))])
X_books = df[allowed_features]
pipeline.fit(X_books)
with open("book_recommender.pkl", "wb") as f:
    pickle.dump(pipeline, f)
```
